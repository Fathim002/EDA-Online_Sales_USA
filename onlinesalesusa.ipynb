### 1.How does the total revenue generated each months?

```python
revenue_by_year = df.groupby('Month')['Total'].sum().sort_values(ascending=False).reset_index()
print(revenue_by_year)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(data=revenue_by_year,x='Month', y='Total', palette='viridis')
plt.title('Total Revenue by Year')
plt.xticks(rotation=90)
plt.show()
```

*December is typically the peak month for online sales in the USA, largely driven by the holiday shopping season. During this time, consumers engage in heightened purchasing activity as they seek gifts for Christmas and other year-end celebrations. The combination of festive promotions, holiday discounts, and the convenience of online shopping leads to significant increases in revenue for e-commerce businesses.*

### 11.How does the presence or absence of a discount affect the total order value?

```python
discounted_orders = df['Discount_amount'] > 0 
percentage_discounted = (discounted_orders.sum() / len(df)) * 100 
print(percentage_discounted)
```

```python
import matplotlib.pyplot as plt

discount_counts = [discounted_orders.sum(), len(df) - discounted_orders.sum()]
labels = ['Discounted Orders', 'Non-Discounted Orders']
colors = ['#66B3FF', '#FFCC99']

plt.figure(figsize=(6, 6))
plt.pie(discount_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Percentage of Discounted vs. Non-Discounted Orders')
plt.show()
```

*Discounts play a significant role in influencing customer purchasing behavior.Discounts not only attract price-sensitive customers but also encourage higher purchase volumes and larger order values, as customers often buy additional items to maximize their savings.Promotional discounts, seasonal sales, and limited-time offers appear particularly effective in boosting sales, creating a sense of urgency and driving impulse purchases.*

### 18.What is the average length of time that customers have been associated with our company?

```python
df["customer_tenure"] = pd.to_datetime("today") - pd.to_datetime(df["Customer_since"])
df["customer_tenure_days"] = df["customer_tenure"].dt.days
df['customer_tenure_days']
```

```python
counts, bin_edges = np.histogram(df["customer_tenure_days"], bins=5)

bin_midpoints = 0.5 * (bin_edges[1:] + bin_edges[:-1])

plt.figure(figsize=(8, 5))
plt.plot(bin_midpoints, counts, marker='o')
plt.xlabel("Customer Tenure (Days)")
plt.ylabel("Number of Customers")
plt.title("Distribution of Customer Tenure")
plt.xticks(bin_edges)
plt.grid()
plt.show()
```

*The average length of time customers have been associated with our company is a key indicator of customer loyalty and satisfaction.This metric reflects how long customers have engaged with the brand, which can influence their purchasing behavior and trust in the company*
